```python
#Calulate the Insta Engagement Rate
#Instagram: Engagement Rate = (Likes+Comments+Shares)/Total Followers X 100
def engagement_rate_instagram(followers, likes, comments, shares):
    return (likes + comments + shares) / followers * 100

#Calulate the Twitter Engagement Rate
#Twitter Engagement Rate = (Likes+Retweets)/Total Followers X 100
def engagement_rate_twitter(followers, likes, retweets):
    return (likes + retweets) / followers * 100
   
#Calulate the Facebook Engagement Rate
#Facebook Engagement Rate = (Reactions+Comments+Shares)/Total Followers X 100
def engagement_rate_facebook(followers, reactions, comments, shares): 
    return ((reactions + comments + shares) / followers) * 100

#Take a param of user_input(str), check if the user_input contains all digit which is positive integer, and return a positive integer.
def validate_positive_integer(prompt):
    while True:
        user_input = input(prompt)
        # Check if input is a number using isdigit() and is greater than 0
        if user_input.isdigit() and int(user_input) > 0:
            return int(user_input)
        print("Invalid input. Please enter a positive integer.")

def validate_platform_string(prompt):
    while True:
        user_input = input(prompt)
        # Check if input is 1 of the followings: Instagram | instagram | Twitter | twitter | Facebook | facebook
        valid_platforms = ['instagram', 'twitter', 'facebook']
        if user_input.lower() in valid_platforms:
            return user_input
        print("Invalid input. Please enter again.")
    
def user_input_and_print():

    # Input and validate if the number_of_posts is positive integer
    number_of_posts = validate_positive_integer('How many posts do you want to analyze? ')
   
    for post in range(int(number_of_posts)):

        print('Post ' + str(post + 1) + ':')

        #Ask user to input platform name
        platform = validate_platform_string('Enter the platform (Instagram, Twitter, Facebook): ')

        # Instagram
        if platform.lower() == 'instagram':
            followers = validate_positive_integer('Enter the total number of followers: ')
            likes = validate_positive_integer('Enter the total number of likes: ')
            comments = validate_positive_integer('Enter the total number of comments: ')
            shares = validate_positive_integer('Enter the total number of shares: ')
            engagement_rate = engagement_rate_instagram(followers, likes, comments, shares)
            formatted_engagement_rate = f'{engagement_rate:.2f}'
            print('The engagement rate for Post ' + str(post + 1) + ' on ' + platform.capitalize() + ' is: ' + formatted_engagement_rate + '%')

        # Twitter
        if platform.lower() == 'twitter':
            followers = validate_positive_integer('Enter the total number of followers: ')
            likes = validate_positive_integer('Enter the number of likes: ')
            retweets = validate_positive_integer('Enter the number of retweets: ')
            engagement_rate = engagement_rate_twitter(followers, likes, retweets)
            formatted_engagement_rate = f'{engagement_rate:.2f}' # Format the engagement rate as a percent
            print('The engagement rate for post ' + str(post+1) + ' on ' + platform.capitalize() + ' is: ' + formatted_engagement_rate + '%') # Print the engagement rate 
            
        # Facebook
        if platform.lower() == 'facebook':
            followers = validate_positive_integer('Enter the total number of followers: ')
            reactions = validate_positive_integer('Enter the total number of reactions: ')
            comments = validate_positive_integer('Enter the total number of comments: ')   
            shares = validate_positive_integer('Enter the total number of shares: ')
            engagement_rate = engagement_rate_facebook(followers, reactions, comments, shares) #Calculate the engagement rate using the function
            formatted_engagement_rate = f'{engagement_rate:.2f}'
            print('The engagement rate for Post ' + str(post + 1) + ' on ' + platform.capitalize() + ' is: ' + formatted_engagement_rate + '%')
                    
# Call the function
user_input_and_print()
```
