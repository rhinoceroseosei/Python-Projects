```python
# Mini-project 2
# List of financial transactions
# Each transaction is a sub-list with two components: description and amount.
transactions = [
    ["Rent payment for office", "1200"],
    ["Electricity bill", "300.50"],
    ["Payroll for March", "8000"],
    ["Printer paper and supplies", "250"],
    ["Gas bill", "150.75"],
    ["Unknown transaction", "XYZ"],  # Invalid entry
    ["Internet bill", "200"]
]

# Categories of transactions
categories = { 
    "Rent": ["rent", "lease"], 
    "Utilities": ["electricity", "water", "gas", "internet"], 
    "Salaries": ["salary", "wages", "payroll"], 
    "Office Supplies": ["paper", "stationery", "printer", "office supplies"], 
    "Miscellaneous": []
}

# Handle processing string amount to float amount, and remove the invalid entries
def handle_str_to_float(transactions: list) -> list:
    processed_list = []
    for transaction in transactions:
        description, amount_str = transaction # list unpacking
        try:
            # Try to convert a string of number to a float
            amount = float(amount_str)

            # Add the valid transaction to the new list
            processed_list.append([description, amount])

        except ValueError:
            # Skip invalid transaction
            continue
    
    return processed_list

# Define a function to categorize transactions based on keywords found in the description. 
# The function takes description as input argument and loops the categories and keywords to assign the correct category.
def categorize_transactions(description: str) -> str:
    description_lower = description.lower()
    
    for category, keywords in categories.items():
        for keyword in keywords:
            if keyword in description_lower:
                return category
    
    return "Miscellaneous"

# Function to calculate the total amount spent in each category
def calculate_total_by_category(transactions: list) -> dict:
    totals = {}
    for category in categories.keys():
        totals[category] = 0

    for description, amount in transactions:
        category = categorize_transactions(description)
        totals[category] += amount

    return totals

# Function to display the summary
def display_summary_report(summary_dictionary: dict):
    print('Detailed Transaction Summary'.center(40, '-'))
    print('-' * 40)
    print('Category'.ljust(20) + 'Total Amount'.rjust(20))
    print('-' * 40)
    
    for category, total in summary_dictionary.items():
        category_adjusted = category.ljust(20)
        amount_adjusted = ('$' + f'{total:.2f}').rjust(20)
        print(f'{category_adjusted}{amount_adjusted}')

# Call the functions to handle processing string amount to float amount, and remove the invalid entries
valid_transactions = handle_str_to_float(transactions)
summary_dict = calculate_total_by_category(valid_transactions)
display_summary_report(summary_dict)
```
